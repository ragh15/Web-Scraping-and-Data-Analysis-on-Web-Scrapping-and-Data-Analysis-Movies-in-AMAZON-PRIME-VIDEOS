```python
page = requests.get("https://www.finder.com/in/amazon-prime-movies")
```

```python
soup = BeautifulSoup(page.text)
```

```python
s=soup.find_all("td")
```

```python
print(s,end='')
```

```python
s=soup.find_all("td")
```

```python
Name=[]
for i,k in enumerate(s):
    a= k.text
    if i%8==0:
        Name.append(a)
Name

#extracting names
```

```python
Rating=[]
for i,char in enumerate(s):
    a = char.text
    if i%8==1:
        Rating.append(a)
Rating
#extacting Rating
        
    
```

```python
soup.find_all("td")
```

```python
Duration=[]
for i in soup.find_all("td"):
    a = i.text
    b = re.findall(r".* min",a)
    if b:
        Duration.append(b[0])
Duration
#extracting Duration
```

```python
s=soup.find_all("td")
s
```

```python
Year=[]
for i,char in enumerate(s):
    a = char.text
    if i%8==3:
        Year.append(a)
Year
#extracting Year
```

```python
s=soup.find_all("td")
s
```

```python
Age=[]
for i,char in enumerate(s):
    a = char.text
    if i%8==4:
        Age.append(a)
Age
#extracting Age
```

```python
s=soup.find_all("td")
s
```

```python
Genre=[]
for i,char in enumerate(s):
    a = char.text
    if i%8==6:
        Genre.append(a)
Genre
#extracting Genre
```

```python
Subtitles=[]
for i,char in enumerate(s):
    a = char.text
    if i%8==7:
        Subtitles.append(a)
Subtitles
#extracting Subtitles
```
